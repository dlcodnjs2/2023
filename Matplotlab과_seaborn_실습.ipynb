**[숙제 3]** 그래프 전체 크기 지정하는 코드 (너비 8인치, 높이 6인치) / 그래프의 제목을 Plot of age vs. Height (in cms)<\n은 줄바꿈>로 설정하는 코드 (글자 크기 20, 글자 스타일 italic으로 설정) / 그래프 가로축과 세로축에 대해 격자무늬 표시해주는 코드 / 그래프 y축의 최솟값을 100, 최댓값을 200으로 설정하는 코드 / 가로축의 눈금을 0~55까지 5단위로 설정하고 가로축과 세로축의 눈금 값을 나타내는 숫자 크기를 15로 설정하는 코드 / 점 (20,100)에서 점 (20,180)을 따라 수직선을 표시해주는 코드 / 점 (0,180)에서 점 (55,180)을 따라 수평선을 표시해주는 코드 / 그래프를 화면에 나타나도록 해주는 코드

```python
# 그래프 전체 크기 지정하는 코드 (너비 8인치, 높이 6인치)
plt.figure(figsize=(8,6))

# 그래프의 제목을 Plot of age vs. Height (in cms)<\n은 줄바꿈>로 설정하는 코드 (글자 크기 20, 글자 스타일 italic으로 설정)
plt.title("Plot of Age vs. Height (in cms)\n",fontsize=20, fontstyle='italic')

# X- and Y-label with fontsize
plt.xlabel("Age (years)",fontsize=16)
plt.ylabel("Height (cms)",fontsize=16)

# 그래프 x축과 y축에 대해 격자무늬 표시해주는 코드
plt.grid (True) 

# 그래프 y축의 최솟값을 100, 최댓값을 200으로 설정하는 코드
plt.ylim(100,200)

# x축의 눈금을 0~55까지 5단위로 설정하고 가로축과 세로축의 눈금 값을 나타내는 숫자 크기를 15로 설정하는 코드
plt.xticks([i*5 for i in range(12)],fontsize=15)
plt.yticks(fontsize=15)

# Main plotting function with choice of color, marker size, and marker edge color
plt.scatter(x=age,y=height,c='orange',s=150,edgecolors='k')

# Adding bit of text to the plot
plt.text(x=15,y=105,s="Height increaes up to around \n20 years and then tapers off",fontsize=15, 
         rotation=30, linespacing=2)
plt.text(x=22,y=185,s="Nobody has a height beyond 180 cm",fontsize=15)

# 점 (20,100)에서 점 (20,180)을 따라 수직선을 표시해주는 코드
plt.vlines(x=20,ymin=100,ymax=180,linestyles='dashed',color='blue',lw=3)

# 점 (0,180)에서 점 (55,180)을 따라 수평선을 표시해주는 코드
plt.hlines(y=180,xmin=0,xmax=55,linestyles='dashed',color='red',lw=3)

# Adding a legend
plt.legend(['Height in cms'],loc=2,fontsize=14)

# 그래프를 화면에 나타나도록 해주는 코드
plt.show()
```

# Simple line plot

**[숙제 6]** days에 1부터 30까지 1씩 증가하는 배열을 저장하고 candidate_A와 candidate_B에 저장하려는 식 안에 공통적으로 들어있는 np.random.randn(30)은 표준정규분포 내에서 0~1사이의 실수값들을 임의로 추출하여 길이가 30인 배열을 만든다. candidate_A와 candidate_B에 각각 주어진 식들을 계산하여 얻은 배열을 각각의 변수에 저장한다는 뜻이다.

```python
days = np.arange(1,31)
candidate_A = 50+days*0.07+2*np.random.randn(30)
candidate_B = 50-days*0.1+3*np.random.randn(30)
```

**[숙제 7]** ymin(y축 최솟값)에 candidate_A에서의 최솟값과 candidate_B에서의 최솟값 중 더 작은 값을 저장하고 ymax(y축 최댓값)에 candidate_A에서의 최댓값과 candidate_B에서의 최댓값 중 더 큰 값을 저장한다는 뜻

```python
# Determine the minimum and maximum of stock prices
ymin = min(candidate_A.min(),candidate_B.min())
ymax = max(candidate_A.max(),candidate_B.max())
```

**[숙제 8-1]** plt.style.use('fivethirtyeight') 의 뜻 : 그래프의 스타일을 'fivethirtyeight'로 지정한다는 뜻 / 
plt.plot(days,candidate_A,'o-',markersize=10,c='blue',lw=2)의 옵션 : days(그래프의 가로축 데이터), candidate_A(그래프의 세로축 데이터), o-(원형마커와 실선), markersize=10(원형마커의 크기를 10으로 설정), c='blue'(그래프의 색을 파란색으로 설정), lw=2(그래프의 선 굵기를 2로 설정)

**[숙제 8-2]**  plt.plot(days,candidate_A,'o-',markersize=10,c='blue',lw=2)에서 ‘o-‘를 ‘s-‘로, markersize=15로, c=’blue’로, lw=5로 바꾸고, 
plt.plot(days,candidate_B,'^-',markersize=10,c='green',lw=2)에서 markersize=15로, c=’y’로, lw=5로 바꿈


```python
# Set style
plt.style.use('fivethirtyeight')

plt.figure(figsize=(12,5))
plt.title("Time series plot of poll percentage over a month\n",fontsize=20, fontstyle='italic')
plt.xlabel("Days",fontsize=16)
plt.ylabel("Poll percentage (%)",fontsize=16)
plt.grid (True)
plt.ylim(ymin*0.98,ymax*1.02)
plt.xticks([i*2 for i in range(16)],fontsize=14)
plt.yticks(fontsize=15)

# Main plotting function - plot (note markersize, lw (linewidth) arguments)
plt.plot(days,candidate_A,'s-',markersize=15,c='orange',lw=5)
plt.plot(days,candidate_B,'^-',markersize=15,c='y',lw=5)

plt.legend(['Poll percentage of candidate A (%)', 'Poll percentage of candidate B (%)'],loc=2,fontsize=14)
plt.show()
```

# Boxplot

```python
plt.style.use('ggplot')


# Note how to convert default numerical x-axis ticks to the list of string by passing two lists 
plt.boxplot(x=[candidate_A,candidate_B],showmeans=True)
plt.grid(True)
plt.xticks([1,2],['Candidate A','Candidate B'])

#plt.yticks(fontsize=15)

plt.show()
```

# Correlation matrix and heatmap

```python
import numpy as np
```

**[숙제 18]** 데이터프레임 df의 모든 열들 간의 상관계수를 계산하여 그 계산하여 나온 행렬을 corr_mat라는 함수에 저장한다는 뜻

```python
corr_mat=np.corrcoef(df,rowvar=False)
```

**[숙제 19]** corr_mat의 크기

```python
corr_mat.shape
```

**[숙제 20]** corr_mat를 데이터프레임으로 변환하여 corr_df에 저장하는데 corr_mat의 열 이름은 df의 열 이름을 그대로 사용하고 corr_mat의 행 이름도 df의 열 이름을 사용한다는 뜻

```python
corr_df = pd.DataFrame(corr_mat,columns=df.columns,index=df.columns)
```

**[숙제 21]** corr_mat라는 상관 계수 행렬의 각 요소를 반올림하여 소수점 3째자리까지만 나타나도록 설정

```python
print(np.round(corr_mat,3))
```
